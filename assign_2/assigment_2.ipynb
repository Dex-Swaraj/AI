```python
import sys

left_side = set()
right_side = set()
farmer_side = "left"

def initialize_game():
    global left_side
    left_side.add("farmer")
    left_side.add("wolf")
    left_side.add("goat")
    left_side.add("cabbage")

def is_safe(side):
    return not (("wolf" in side and "goat" in side and "farmer" not in side) or
                ("goat" in side and "cabbage" in side and "farmer" not in side))

def move_item(item):
    global farmer_side, left_side, right_side
    if farmer_side == "left":
        if item is not None:
            if item not in left_side:
                return "Invalid move!"
            left_side.remove(item)
            right_side.add(item)
        farmer_side = "right"
        left_side.remove("farmer")
        right_side.add("farmer")
    else:
        if item is not None:
            if item not in right_side:
                return "Invalid move!"
            right_side.remove(item)
            left_side.add(item)
        farmer_side = "left"
        right_side.remove("farmer")
        left_side.add("farmer")

    if not is_safe(left_side) or not is_safe(right_side):
        return "Move not allowed! Items left alone unsafely."
    return "Move successful!\n"

def display_state():
    print("Left Side:", ", ".join(left_side))
    print("Right Side:", ", ".join(right_side))
    print("Farmer is on the", farmer_side, "side.")

def is_game_won():
    return len(left_side) == 0

def main():
    initialize_game()
    print("Welcome to the Wolf, Goat, and Cabbage game!")
    print()
    display_state()

    while not is_game_won():
        move_input = input("What do you want to move? (wolf, goat, cabbage, none): ").strip().lower()
        if move_input == "none":
            item = None
        elif move_input not in ["wolf", "goat", "cabbage"]:
            print("Invalid item! Please choose wolf, goat, cabbage, or none.")
            continue
        else:
            item = move_input

        result = move_item(item)
        print(result)
        display_state()

        if result.startswith("Move not allowed"):
            break

    if is_game_won():
        print("Congratulations! You won the game!")
    else:
        print("Game over! Try again.")

if __name__ == "__main__":
    main()
```
